```python
df =pd.read_excel("/content/withoutEmployeeID.xlsx")
print(df['Date'][0])
type(df["Date"])
```

```python
# convert it to a date-time
df["Date"] = pd.to_datetime(df["Date"], format = '%d/%m/%Y')
# display the first date
df["Date"][0]
```

```python
df['year'] = pd.DatetimeIndex(df['Date']).year
df['month'] = pd.DatetimeIndex(df['Date']).month
df['day'] = pd.DatetimeIndex(df['Date']).day
NewDfWithoutDateColumn=df.drop('Date', axis=1)
NewDfWithoutDateColumn.head()
```

```python
# extracting the Season out of the date
NewDfWithoutDateColumn['Season']=NewDfWithoutDateColumn['month']

season_mapping = {
                  12: 'Winter',1: 'Winter',2: 'Winter',
                   3: 'Spring', 4: 'Spring', 5: 'Spring',
                   6: 'Summer', 7: 'Summer', 8: 'Summer',
                   9: 'Autumn', 10: 'Autumn', 11: 'Autumn'
                 }

NewDfWithoutDateColumn['Season'] = NewDfWithoutDateColumn['Season'].map(season_mapping)
NewDfWithoutDateColumn.head()
```

```python
# extracting the Month out of the date
NewDfWithoutDateColumn['Name_of_Month']=NewDfWithoutDateColumn['month']

season_mapping = {
                  1: 'January', 2:'February', 3:'March', 4:'April', 5:'May', 6:'June',
                  7:'July', 8:'August', 9:'September', 10:'October', 11:'November',12:'December'
                 }

NewDfWithoutDateColumn['Name_of_Month'] = NewDfWithoutDateColumn['Name_of_Month'].map(season_mapping)
NewDfWithoutDateColumn.head()
```

```python
# extracting the Weekday out of the date
# Convert the date column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

NewDfWithoutDateColumn['Weekday'] =df['Date'].dt.day_name()
NewDfWithoutDateColumn.head()
```

2.1. توجه داشته باشید که فقط مقادیر 1 تا 21 کدهای موجود در جدول ICD را نشان میدهد.از آنجایی که تمام مقادیر موجود در ICD نشاندهنده یک نوع بیماری است،
یک متغیر binary جدید تحت عنوان Disease تعریف کنید که بررسی کند دلیل غیبت  برای هر شخص در بازه ICD قرار دارد یا خیر.
</div>

```python
def in_icd(val):
  if val<=21:
    return 'ICD'
  else:
    return 'not ICD'

NewDfWithoutDateColumn["Disease"] = NewDfWithoutDateColumn["Absence_Reason"].apply(in_icd)
print(NewDfWithoutDateColumn["Disease"])
```

سپس هر دو نمودار بالا را مجددا رسم کنید و با تحلیل این این نمودارها درمورد تاثیرگذاری این دو فاکتور بر روی زمان غیبت توضیح دهید.

```python
Distance_array , m = stats.yeojohnson(df['Distance_to_Work'].to_numpy())
Duration_array , n = stats.yeojohnson(df['Absence_Duration_Hours'].to_numpy())

sns.jointplot( x=Distance_array , y=Duration_array)
plt.ylabel('Absence Duration (hours)')
plt.xlabel('Distance_to_Work')
plt.show()
```

```python
Expense_array , m = stats.yeojohnson(df['Transportation_Expense'].to_numpy())
Duration_array , n = stats.yeojohnson(df['Absence_Duration_Hours'].to_numpy())

sns.jointplot( x=Expense_array , y=Duration_array)
plt.ylabel('Absence Duration (hours)')
plt.xlabel('Transportation Expense')
plt.show()
```
